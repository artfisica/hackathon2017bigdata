```python
import ROOT
```

```python
%jsroot on
```

```python
##f = ROOT.TFile.Open("http://universidad.ch/hackathon2017geneva/data/root/AGSEC10.root")
f = ROOT.TFile.Open("AGSEC10.root")
```

```python
c = ROOT.TCanvas("testCanvas","a first way to plot a variable",750,500)
c.SetGrid()
```

```python
t = f.Get("data")
```

```python
h1 = ROOT.TH1F("variable1","Number of equipment of people that do rent other",24,0,24)
h1.SetLineColor(2)
```

```python
h2 = ROOT.TH1F("variable2","Number of equipment of people that do not rent other",24,0,24)
h2.SetLineColor(3)
```

```python
for event in t:
    if t.a10q6==1:
        h1.Fill(t.A10itemcod)
    if t.a10q6==2:
        h2.Fill(t.A10itemcod)

print "Done!"
```

```python
h1.Draw()
c.Draw()
```

```python
h2.Draw()
c.Draw()
```

percentage of people that rent or borrow equipment

```python
percentage = (h1.Integral()/(h1.Integral()+h2.Integral()))*100
print percentage
```

```python
scale1 = h1.Integral()
h1.Scale(1/scale1)
h1.Draw()
c.Draw()
```

```python
scale2 = h2.Integral()
h2.Scale(1/scale2)
h2.Draw()
c.Draw()
```

```python
h2.Divide(h1)
h2.SetLineColor(4)
##
line = ROOT.TLine(0,1,24,1)
line.SetLineColor(2)
line.SetLineWidth(3)
h2.Draw("ep")
line.Draw()
c.Draw()
```
